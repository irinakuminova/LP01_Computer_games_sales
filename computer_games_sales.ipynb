## Подготовка данных

```python
# импортируем библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st
```

```python
# делаем настройки отображения
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('/datasets/games.csv')
df.head()
```

```python
df.columns = df.columns.str.lower() #приводим названия столбцов к нижнему регистру
df.head()
```

```python
df['platform'].unique() # смотрим список уникальных значений столбца 'platform'
```

Дополним данные информацией о компаниях-производителях игровых платформ.

```python
sony = pd.DataFrame({'company': 'sony', 'platform': ['PS', 'PS2', 'PS3', 'PS4', 'PSV', 'PSP', 'PC']})
ms = pd.DataFrame({'company': 'ms', 'platform': ['X360', 'XOne', 'XB']})
nintendo = pd.DataFrame({'company': 'nintendo', 'platform': ['GB', 'DS', 'NES', 'SNES', 'GBA', '3DS', 'N64', 'Wii', 'WiiU']})
others = pd.DataFrame({'company': 'others', 'platform': ['2600', 'GC', 'GEN', 'DC', 'SAT', 'SCD', 'WS', 'NG', 'TG16', '3DO', 'GG', 'PCFX']})
company = pd.concat([sony, ms, nintendo])
company
```

```python
df = pd.merge(df, company, how='left', on='platform')
df.head()
```

Приведем столбец 'genre' к нижнему регистру.

```python
df['genre'] = df['genre'].str.lower()
df['genre'].unique()
```

Приведем столбец 'name' к нижнему регистру.

```python
df['name'] = df['name'].str.lower()
```

```python
df.info()
```

Посмотрим содержимое строк, для к-ых отсуствуют значения в столбце 'name'.

```python
df[df['name'].isnull()]
```

В строках с пустым 'name' много отсуствующих данных, поэтому удаляем эти строки.

Также удаляем строки с пустым 'year_of_release' и переводим данные в целочисленный формат.

```python
df.dropna(subset=['name'], inplace=True)

df.dropna(subset=['year_of_release'], inplace=True)
df['year_of_release'] = df['year_of_release'].astype('int')
df.info()
```

Пропуски в 'critic_score' и 'user_score' заменим условным маркером -1, потому это это субъективная оценочная категория, которую невозможно рассчитать.

```python
df['critic_score'] = df['critic_score'].fillna(-1) #пропуски в 'critic_score' заменяем на условный маркер -1
```

```python
df['user_score'] = df['user_score'].fillna(-1) #пропуски в 'user_score' заменяем на условный маркер -1
df['user_score'].unique()
```

```python
df[df['user_score'] == 'tbd'] #в 'user_score' встречается значение 'tbd', выводим на экран эти строки
```

Пропуски в 'tbd' в 'user_score' также заменим условным маркером -1, поскольку вариант оценки невозможно рассчитать.

```python
#'tbd' - to be determened, т.е. рейтинг не может быть определен из-за малого кол-ва данных
df['user_score'] = np.where(df['user_score'] == 'tbd', -1, df['user_score']) #заменим 'tbd' также на условный маркер -1
df['user_score'] = df['user_score'].astype('float') #данные в 'user_score' переводим в числовой формат float
```

```python
df.info()
```

```python
df['rating'].unique() #выводим список уникальных значений в столбце 'rating'
```

Рейтинги ESRB действительны в США и Канаде. В Европе и Японии существуют свои рейтинговые системы оценки игр: PEGI в Европе, CERO в Японии. 
Поэтому отсутствующий рейтинг не всегда означает, что он ожидается, и можно применить замену на RP (Rating Pending — Рейтинг ожидается). Отсутствующий рейтинг может означать в том числе, что игра проходила оценку по другой рейтинговой системе, поэтому замена на маркер 'unk' (unknown) будет более полезной в данном случае.

```python
df['rating'] = df['rating'].fillna('unk')
```

```python
df.loc[df.duplicated(subset=['name', 'platform','genre','year_of_release'], keep=False)] #найдем полные дубликаты
```

```python
df = df.drop_duplicates(subset=['name', 'platform','genre','year_of_release'], keep='first').copy() #удалим добликат
```

Добавим в таблицу новый столбец 'total_sales', просуммировав для него значения столбцов с 'na_sales' по 'other_sales'.    

```python
df['total_sales'] = df.loc[:, 'na_sales':'other_sales'].sum(axis=1)
df.head()
```

**Вывод**

В ходе подготовки данных названия столбцов были приведены к нижнему регистру. Также к нижнему регистру приведены значения в столбцах 'genre' и 'name', чтобы избавится от дубликатов из-за использования разных регистров (если такое имелось).

В столбце 'year_of_release' тип данных float изменен на целочисленный int.

В столбце 'user_score' после замены пропусков и строковых значений тип данных object изменен числовой float.

Строки с пропусками в столбце 'name' (2 шт) удалены, поскольку в них отсутствовали остальные ключевые параметры - платформа и жанр.

Строки с пропусками в 'year_of_release' (269 шт - 1.6%) удалены, поскольку нет варианта замены пропусков достоверными данными. 

Пропуски в 'critic_score' и 'user_score' заменены условным маркером -1, чтобы иметь возможность фильтровать по нему. Отсутствие значений говорит скорее всего о том, что игра не имеет достаточно оценок, чтобы определить рейтинг.
Строковое значение 'tbd' (to be determined) в столбце 'user_score' также заменено на маркерное значение -1, поскольку говорит о том же самом - недостаточно оценок, чтобы определить рейтинг.

Пропуски в 'rating' заменены на RP («Rating Pending» — «Рейтинг ожидается»).

Были обнаружены две полностью задублированные строки, дубликат был удален.

После предобработки данных в датасет добавлены столбецы: 
- 'total_sale' с суммарными продажами по всем регионам,
- 'company' с производителем платформы.

### Динамика продаж по платформам

Посмотрите, как менялись продажи по платформам. Для этого выберем платформы с наибольшими суммарными продажами (топ-10) и построим распределение по годам.

```python
platforms_top10 = (df
                    .pivot_table(index='platform', values='total_sales', aggfunc='sum')
                    .sort_values(by='total_sales', ascending=False)
                    .reset_index()
                    .head(10)
                    )
platforms_top10 #список 10 платформ с максимальными продажами
```

```python
df_top10 = pd.concat(df[df['platform'] == y] for y in platforms_top10['platform'].unique())
df_top10.head() #выборка из общего датафрейма данных по платформам из топ-10
```

```python
df_top10_grouped = (df_top10
                    .pivot_table(index=(['platform', 'year_of_release']), values='total_sales', aggfunc='sum')
                    .reset_index()
                    )
df_top10_grouped #суммарные продажи по годам для каждой платформы из топ-10
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x='year_of_release', y='total_sales', data=df_top10_grouped, hue='platform', legend='brief')

plt.title('Динамика продаж по годам по 10 популярным платформам', fontsize=16)
plt.xlabel('years', fontsize=12)
plt.ylabel('total_sales', fontsize=12)
plt.grid()
plt.show()
```

**Вывод**

По данным видно, что из топ-10: 5 платформ - это платформы PlayStation от Sony (PS, PS2, PS3, PS4, PSP), 4 платформы от Nintendo (3DS, DS, GBA, Wii), и 1 платформа от MS ({360).
    
По графику видно, что средний срок жизни платформы - около 10 лет. При этом по группе PS наблюдается, что каждый следующий релиз выходит примерно через 6 лет после выхода предыдущего, и в течении оставшихся 4-х лет новый релиз полностью замещает предыдущий.

### Лидеры продаж за 2013-2016 гг

Для прогноза продаж на 2017 г. будем использовать данные с 2013 г., поскольку это период показывает ситуацию на рынке компьютерных игр при наличии его главного конкурента - рынка мобильных и планшетных игр.

### Отзывы и продажи

Посмотрите, как влияют на продажи внутри самой популярной платформы PS4 от Sony отзывы пользователей и критиков. Для этого построим диаграммы рассеивания и посчитаем корреляцию между отзывами и продажами.

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 4))
plt.suptitle('Диаграммы рассеивания для платформы PS4', size=14)

pad = 2
title_size = 11
sns.scatterplot(data=df.query('critic_score != -1 & platform == "PS4"'), #зависимость продаж от отзывов критиков
                x='critic_score',
                y='total_sales',
                ax=ax[0],
                color='#0B5394'
                )

sns.scatterplot(data=df.query('user_score != -1 & platform == "PS4"'), #зависимость продаж от отзывов пользователей
                x='user_score',
                y='total_sales',
                ax=ax[1],
                color='#38761d'
               )

plt.show()
```

```python
#корреляция продаж и отзывов критиков
(df.query('critic_score != -1 & platform == "PS4"')['total_sales']
  .corr(df.query('critic_score != -1 & platform == "PS4"')['critic_score'])
)
```

```python
#корреляция продаж и пользовательских отзывов
(df.query('user_score != -1 & platform == "PS4"')['total_sales']
  .corr(df.query('user_score != -1 & platform == "PS4"')['user_score'])
)
```

**Вывод**

Полученные данные показывают, что пользовательские отзывы совершенно не влияют на продажи игр на платформе PS4. Отзывы критиков оказывают слабое влияние, и их тоже можно не принимать во внимание.

Посмотрим на остальные платформы и зависимость их продаж от отзывов. Будем рассматривать период продаж с 2013 г.

## Проверка гипотез

### Сравнение пользовательских рейтингов платформ XOne и PC

**Выдвинем гипотезу №1:** средние пользовательские рейтинги платформ XOne и PC одинаковые. 

Нулевая гипотеза будет звучать аналогично. Альтернативная: средние рейтинги не равны.

Для ее проверки воспользуемся мнетодом scipy.stats.ttest_ind (array1, array2, equal_var).
Зададим пороговое значение на уровне 5%.

```python
alpha = 0.05 # критический уровень статистической значимости

xone = df.query('user_score != -1 & year_of_release >= 2013 & platform == "XOne"')['user_score']
pc = df.query('user_score != -1 & year_of_release >= 2013 & platform == "PC"')['user_score']

results_platforms = st.ttest_ind(xone, pc, equal_var=False)

print('p-значение:', results_platforms.pvalue)

if results_platforms.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

**Вывод**

Получили высокую вероятность ошибиться (p-value 14.8%) при отклонении нулевой гипотезы. Значит можно говорить о том, что средние пользовательские оценки XOne и PC одинаковы.

### Сравнение пользовательских рейтингов жанров Action и Sports

**Выдвинем гипотезу №2:** средние пользовательские рейтинги жанров Action и Sports разные. 
Для этой гипотезу сформулируем нулевую гипотезу, которая будет звучать так: средние пользовательские рейтинги жанров Action и Sports одинаковые.
Для ее проверки воспользуемся мнетодом scipy.stats.ttest_ind (array1, array2, equal_var).
Зададим пороговое значение на уровне 5%.

```python
alpha = 0.05 # критический уровень статистической значимости

action = df.query('user_score != -1 & year_of_release >= 2013 & genre == "action"')['user_score']
sports = df.query('user_score != -1 & year_of_release >= 2013 & genre == "sports"')['user_score']

results_genres = st.ttest_ind(action, sports, equal_var=False)

print('p-значение:', results_genres.pvalue)

if results_genres.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

**Вывод**

Получили очень низкую вероятность ошибиться при отклонении нулевой гипотезы. Значит можно говорить о том, что средние пользовательские оценки жанров Action и Sports различаются.

Японский рынок и здесь выделяется, тем, что почти половина игр, выпускаемая в этом регионе, не имеют рейтинга по ESRB, поскольку скорее всего пользуются своей рейтинговой системой.


**Рекомендации**

Делая прогноз на 2017 г., можно предположить, что есть вероятность дальнейшего падения глобальных продаж компьютерных игр. 

На североамериканском и европейских рынках основную долю продаж следует ожидать от платформ PS4 и XOne, поскольку предыдущие версии этих платформ скорее всего полностью ушли с рынка, а возможно, и с поддержки их со стороны производителя, а новым еще рано появляться. На японском рынке 3DS от Nintendo и разные версии платформы Play Station скорее всего также возьмут на себя весь объем продаж в регионе.

По жанрам на североамериканском и европейском рынках необходимо делать упор на продажи shooter, sports, action, role-playing. На рынке Японии основной доход ожидаемо принесут игры жанра action и role-playing.

По ESRB рейтингам на североамериканском и европейском рынках необходимо принимать во внимание, что 60% рынка приходится на игры с рейтингом М (17+) и Е (для всех). Для Японского рынка нужно анализировать их собственную рейтинговую систему, поскольку почти половина игр, продаются в этом регионе, не имеют рейтинга по ESRB.

Планируя продажи на 2017 г. не имеет смысла ориентироваться на пользовательские рейтинги и рейтинги критиков, их влияние на продажи очень слабо. На рынке есть более весомый фактор, скорее всего, в виде мобильных игра, который и определяет тенденции развития рынка компьютерных игр.
